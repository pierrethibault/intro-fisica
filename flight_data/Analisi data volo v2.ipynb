```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib widget
```

```python
# Read data
d = pd.read_csv('GPS_flight_data.csv')
```

```python
d.columns
```

```python
# Raggio della Terra in metri
R = 63781Card 1 of 1200

# Time
t = d['t']

# Y
y = np.array(R * d['lat'] * np.pi / 180)

# X
x = np.array(R * d['lon'] * np.cos(np.pi * d['lat'] / 180)  * np.pi / 180)

# Z
z = np.array(d['z'])
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(5,4), dpi=100)
ax.plot(d['t'], z, 'k-')
ax.set_xlabel('Tempo (s)')
ax.set_ylabel('Altitudine (m)')
ax.set_title("Dati decollo")
fig.tight_layout()
```

```python
# (rough) Average velocity
v_av = (11600 - 250) / (1335-300)
```

```python
vx = x[1:]-x[:-1]
vy = y[1:]-y[:-1]
vz = z[1:]-z[:-1]
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(6,4), dpi=100)
ax.plot(t[:-1], 3.6*vx, label='Velocita X (km/h)')
ax.plot(t[:-1], 3.6*vy, label='Velocita Y (km/h)')
ax.legend()
```

```python
v = np.sqrt(vx**2 + vy**2 + vz**2)
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(6,4), dpi=100)
plt.plot(t[:-1], 3.6*v)
ax.set_xlabel('Tempo (s)')
ax.set_ylabel('Velocit (km/h)')
ax.set_title('Modulo della velocit');
```
